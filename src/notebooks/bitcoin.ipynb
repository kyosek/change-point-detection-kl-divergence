# Bitcoin

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Load the data
df = pd.read_excel('/Users/kyosuke.morita/project_v2/bitcoin/resources/data/bitcoin_day.xlsx')
df.head()
```

```python
# Transform
df.columns = ['date','open','high','low','close','volumn','marketCap']
df['date'] = pd.to_datetime(df['date'])
df['log_close'] = np.log(df.close)
```

```python
df.head()
```

```python
# Plot the data
sns.lineplot(y='close',x='date',data=df)
plt.title('Bitcoin Daily Close Price (all time)')
plt.show()
```

```python
sns.lineplot(y='log_close',x='date',data=df)
plt.title('Bitcoin Daily Close Price (all time)')
plt.show()
```

```python
sns.lineplot(y='close',x='date',data=df[df['date']>= '2019-01-01'])
plt.title('Bitcoin Daily Close Price (2019 ~)')
plt.show()
```

```python
sns.lineplot(y='log_close',x='date',data=df[df['date']>= '2019-01-01'])
plt.title('Bitcoin Daily Close Price (2019 ~)')
plt.show()
```

```python
sns.lineplot(y='close',x='date',data=df[df['date']< '2017-01-01'])
plt.title('Bitcoin Daily Close Price (~ 2017)')
plt.show()
```

```python
sns.lineplot(y='log_close',x='date',data=df[df['date']< '2017-01-01'])
plt.title('Bitcoin Daily Close Price (log value) (~ 2017)')
plt.show()
```
